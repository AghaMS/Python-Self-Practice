This code plots a hypothetical network to identify hubs and bridges 

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_nodes_from([
    ('A', {"color": "red"}),
    ('B', {"color": "blue"}),
    ('C', {"color": "red"}),
    ('D', {"color": "blue"}),
    ('E', {"color": "blue"}),
    ('F', {"color": "blue"}),
    ('G', {"color": "blue"}),
    ('H', {"color": "blue"})
])
G.add_edges_from([('A','B'),
                  ('A','C'),
                  ('B','C'),
                  ('C','D'),
                  ('C','E'),
                  ('D','E'),
                  ('D','F'),
                  ('D','G'),
                  ('D','H'),
                  ('E','F'),
                  ('G','H'),
                  ('F','G')])
```

```python
color_map = nx.get_node_attributes(G, "color")

for key in color_map:
    if color_map[key] == "red":
        color_map[key] = "pink"
    else:
        color_map[key] = "yellow"
    
node_colors = [color_map.get(node) for node in G.nodes()]
```

```python
size_map = nx.get_node_attributes(G, "color")

for key in size_map:
    if color_map[key] == "red":
        size_map[key] = 900
    else:
        size_map[key] = 550
        
tenure_size = [size_map.get(node) for node in G.nodes()]
```

```python
#pos = nx.circular_layout(G) 
pos = nx.shell_layout(G)
#options = {"node_size": 10, "alpha": 0.8}
#nx.draw_networkx(G, pos, nodelist=['A'], node_color="r")#, **options)
#nx.draw_networkx(G, pos, node_color="b")#, **options)

nx.draw_networkx(G, pos, node_color = node_colors,  node_size = tenure_size)#

plt.show()
```
